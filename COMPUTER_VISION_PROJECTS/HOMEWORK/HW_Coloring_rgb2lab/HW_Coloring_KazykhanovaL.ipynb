# CV-12. Computer Vision: Color vision
## Color representation of the image

TASK

* Change original image brightness value by 30.
* Remove blue-yellow tones from the original image.

`To Zip final images`

```python
# Import libs
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import matplotlib.pylab as plt
```

```python
# Look at the original image
image_original = imread('photo_hawaii.jpg')

plt.figure(figsize=(10,10))
plt.imshow(image_original), plt.axis('off'), plt.title('Original image', size=20)
plt.show()
```

***

## Image brightness

```python
# Convert to lab format
image_lab_0 = rgb2lab(image_original)
```

```python
"""
Change original image brightness by 30
"""
image_lab_0[...,0] = 30
```

```python
"""
Getting new images with brightness +30,
convert image to RGB from Lab format
"""
image_final_0 = lab2rgb(image_lab_0)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_final_0), plt.axis('off'), plt.title('Image with high brightness', size=20)
# plt.show()
plt.savefig('Image with high brightness.png')
```

***

## Remove yellow-blue tones

```python
# Convert to lab format
image_lab_1 = rgb2lab(image_original)
```

```python
"""
Remove yellow-blue tones (b-channel)
"""
image_lab_1[...,2] = 0
```

```python
"""
Getting new images without yellow-blue tones,
convert image to RGB from Lab format
"""
image_final_1 = lab2rgb(image_lab_1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_final_1), plt.axis('off'), plt.title('Image without blue-yellow tones', size=20)
# plt.show()
plt.savefig('Without blue-yellow tones.png')
```

***

## Image with high brightness and without blue-yellow tones

```python
# Convert to lab format
image_lab_3 = rgb2lab(image_original)
```

```python
# Change original image brightness by 30
image_lab_3[...,0] = +30

# Remove yellow-blue tones (b-channel)
image_lab_3[...,2] = 0
```

```python
"""
Getting new images with brightness +30 and without yellow-blue tones,
convert image to RGB from Lab format
"""
image_final_3 = lab2rgb(image_lab_3)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(image_original), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(image_final_3), plt.axis('off'), plt.title('Image with all conditions', size=20)
plt.show()
# plt.savefig('original_and_result_pictures.png')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_final_3), plt.axis('off'), plt.title('Image with all conditions', size=20)
plt.savefig('final_image_with_all_conditions.png')
```
